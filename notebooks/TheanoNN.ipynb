```python
class GNeuralNetwork(nnet.AbstractNeuralNetworkClassifier):
    def prepare(self):
        assert len(self.layers) == 3 and self.layers[2] == 1
        n1, n2, n3 = self.layers
        W0 = theano.shared(value=numpy.random.normal(size=[n2, n1]).astype(floatX), name='W0')
        W1 = theano.shared(value=numpy.random.normal(size=[n2, n1]).astype(floatX), name='W1')
        W2 = theano.shared(value=numpy.random.normal(size=[n3, n2]).astype(floatX), name='W2')
        self.parameters = {'W0': W0, 'W1': W1, 'W2': W2}

        def activation(input):
            first = T.nnet.sigmoid(T.dot(W0, input)) * T.nnet.sigmoid(T.dot(W1, input))
            return T.nnet.sigmoid(T.dot(W2, first))
        return activation
```

```python
class TanhNeuralNetwork(nnet.AbstractNeuralNetworkClassifier):
    def prepare(self):
        assert len(self.layers) == 3 and self.layers[2] == 1
        n1, n2, n3 = self.layers
        W1 = theano.shared(value=numpy.random.normal(size=[n2, n1]).astype(floatX), name='W1')
        W2 = theano.shared(value=numpy.random.normal(size=[n3, n2]).astype(floatX), name='W2')
        self.parameters = {'W1': W1, 'W2': W2}

        def activation(input):
            first = T.tanh(T.dot(W1, input))
            return T.nnet.sigmoid(T.dot(W2, first))
        return activation
```

```python
class PlainNeuralNetwork(nnet.AbstractNeuralNetworkClassifier):
    def prepare(self):
        assert len(self.layers) == 3 and self.layers[2] == 1
        n1, n2, n3 = self.layers
        W1 = theano.shared(value=numpy.random.normal(size=[n2, n1]).astype(floatX), name='W1')
        W2 = theano.shared(value=numpy.random.normal(size=[n3, n2]).astype(floatX), name='W2')
        self.parameters = {'W1': W1, 'W2': W2}

        def activation(input):
            return T.dot(W2, T.tanh(T.dot(W1, input))))
        return activation
```

```python
# Pairwise layer
class PWNeuralNetwork(nnet.AbstractNeuralNetworkClassifier):
    def prepare(self):
        assert len(self.layers) == 3 and self.layers[2] == 1
        n1, n2, n3 = self.layers
        W1 = theano.shared(value=numpy.random.normal(size=[n2, n1]).astype(floatX), name='W1')
        W2 = theano.shared(value=numpy.random.normal(size=[n2, n2, 1]).astype(floatX), name='W2', 
                           broadcastable=[False, False, True])
        self.parameters = {'W1': W1, 'W2': W2}
        
        def activation(input):
            first = T.tanh(T.dot(W1, input))
            second = first.reshape([n2, 1, input.shape[1]]) * first.reshape([1, n2, input.shape[1]])
            return T.nnet.sigmoid(T.sum(W2 * second, axis=[0, 1]))

        return activation
```

```python
# pairwise layer on softmax
class PWSMNeuralNetwork(nnet.AbstractNeuralNetworkClassifier):
    def prepare(self):
        n1, n2, n3 = self.layers
        W0 = theano.shared(value=numpy.random.normal(size=[n2, n1]).astype(floatX), name='W1')
        W1 = theano.shared(value=numpy.random.normal(size=[n2, n1]).astype(floatX), name='W1')
        W2 = theano.shared(value=numpy.random.normal(size=[n2, n2, 1]).astype(floatX), name='W2', 
                           broadcastable=[False, False, True])
        self.parameters = {'W0': W0, 'W1': W1, 'W2': W2}
        
        def activation(input):
            first0 = T.nnet.softmax(T.dot(W0, input))
            first1 = T.nnet.softmax(T.dot(W1, input))
            second = first0.reshape([n2, 1, input.shape[1]]) * first1.reshape([1, n2, input.shape[1]])
            return T.nnet.sigmoid(T.sum(W2 * second, axis=[0, 1]))

        return activation
```

```python
class MultiLayerMultiOutputNN(nnet.MultiLayerNetwork):
    def _prepare(self):
        activation = self.prepare()
        loss_ = lambda x, y, w: self.loss(y, activation(x), w)
        x = T.matrix('X')
        y = T.matrix('y')
        w = T.matrix('w')
        self.Activation = theano.function([x], activation(x))
        self.Loss = theano.function([x, y, w], loss_(x, y.T, w.T))
        for name, param in self.parameters.iteritems():
            self.derivatives[name] = theano.function([x, y, w], T.grad(loss_(x, y.T, w.T), param))
```

```python
nn = MultiLayerMultiOutputNN(layers=[subdata.shape[1], 100, sub_y.shape[1]], loss=nnet.log_loss)
```
