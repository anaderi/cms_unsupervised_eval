# Simple gradient boosting
from lhcb_trigger_ml

```python
import gradient_boosting as gb
from reports import ClassifiersDict
```

```python
clfs = ClassifiersDict()
# clfs['gb'] = GradientBoostingClassifier(subsample=0.1, max_depth=8, max_features=8, learning_rate=0.05, min_samples_leaf=100)
clfs['gb'] = gb.GradientBoosting(loss=gb.AdaLossFunction(), subsample=0.15, max_depth=12, n_estimators=400,
                                 max_features=8, learning_rate=0.05, min_samples_leaf=500, update_tree=True, update_on='all')
```

```python
clfs.fit(trainX, trainY, sample_weight=trainW)
pred = clfs.test_on(testX, testY,  sample_weight=testW)
```

# MatrixNet

```python
from cern_utils import ef_classifier, data_storage
```

```python
mn = ef_classifier.ClassifierEF(directory='mn', formula_name='lev_mn_train', 
                                dataset_name='lev_mn', url_base='test02cern-i.vs.os.yandex.net')
mn.set_params(features=variables, iterations=10000, sync=False)
mn.fit(data_storage.DataStorageDF(trainX.iloc[trainY == 1]), data_storage.DataStorageDF(trainX.iloc[trainY == 0]), 
       trainW[trainY == 1], trainW[trainY == 0])
```

```python
prob_mn = mn.predict(data_storage.DataStorageDF(testX))
```

```python
figure(figsize=[10, 7])
fpr, tpr, _ = roc_curve(testY, prob_mn, sample_weight=testW)
roc = roc_auc_score(testY, prob_mn, sample_weight=testW)
plot(fpr, tpr, label='MN (area = %.3f)' % roc)
legend(loc='lower left')
grid()
```
