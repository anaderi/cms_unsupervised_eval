# This is unsupervised learning test
We are trying to classify the data using only clustering.

```python
import root_numpy
import os
import pandas
from sklearn import cluster
from sklearn import metrics, neighbors
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from IPython.display import display_html
import utils
```

```python
def print_heading(x, level=3):
    display_html("<h%i>%s</h%i>" % (level, str(x), level), raw=True)
```

```python
label_names = ['ttbar', 'Wjets', 'tchan', 'tWchan', 'schan', 'DY', 'WW', 'WZ','ZZ']
```

```python
variables = ["log(Pt_BJ1)", "log(Pt_BJ2)", "log(Pt_J1J2)", "log(Pt_JNotBest)",
             "log(Ht)", "DR_J1J2", "Eta_LJ", "log(M_J1J2)", "log(M_JNotBest)",
             "log(M_JW)", "log(Pttop_BJ1)", "Cos_LepLJ_BJ1", "Cos_WLJ_BJ1",
             "log(Pt_Lep)", "Planarity", "log(Mtop_BJ1)", "Eta_Lep", "N_BJ"]

all_variables = variables + ['weight']
```

```python
folder = '/mnt/w76/notebook/datasets/doudko/samples/'
```

```python
file_labels = {}
for name in os.listdir(folder):
    for i, label in enumerate(label_names):
        if label in name:
            file_labels[name] = i
```

```python
data_parts = []
data_labels = []
data_weights = []
for filename, label in file_labels.iteritems():
    data = root_numpy.root2array(folder + filename, treename='Vars', branches=all_variables)
    data = pandas.DataFrame(data)
    data_labels.append(numpy.zeros(len(data), dtype=int) + label)
    data_weights.append(data['weight'])
    data_parts.append(data.drop(['weight'], axis=1))
    
data = pandas.concat(data_parts, ignore_index=True)
labels = numpy.concatenate(data_labels)
weights = numpy.concatenate(data_weights)
```

```python
figure(figsize=(16, 5))
subplot(121), title('n_events'),     hist(labels, bins=len(label_names) )
subplot(122), title('total weight'), hist(labels, bins=len(label_names), weights=weights)
show()
```

# Selecting train part of events
* we cannot use the weights in training
* processing all the data takes much time

So we generate subset according to the weights, the named subset is used in clustering

```python
p = numpy.array(weights, dtype=float64)
p /= sum(p)
mask = numpy.random.choice(a=range(len(data)), size=int(0.2 * len(data)), p=p, replace=False)
train_mask = mask[:len(mask) // 2 + 10]
trainX = data.iloc[train_mask, :]
trainY = labels[train_mask]
trainW = weights[train_mask]

test_mask = mask[len(mask) // 2 + 10:]

testX = data.iloc[test_mask, :]
testY = labels[test_mask]
testW = weights[test_mask]
```

# Sorted distance to the i-th closest event
i=0 is event itself, the distance to it is always zero. We do not make any difference between features

We are using here simple euclidean metrics (metric tensor is identity).

```python
n_events = 40000
nn = neighbors.NearestNeighbors().fit(testX[: n_events])
distances, _ = nn.kneighbors(testX[:n_events], n_neighbors=15)

for i in [0, 1, 5, 10, 14]:
    plot(numpy.sort(distances[:,  i]), label=str(i))
legend()
```

## Helpful function, prints report on clusterization

```python
def clusterization_report(y_true, y_pred):
    """
    y_true - real labels of events,
    y_pred - predicted number of events
    """
    print_heading("Size of classes in test and predicted classes")
    
    pylab.figure(figsize=(14, 5))
    subplot(121)
    hist(y_true, bins=len(numpy.unique(y_true))), title("Distribution over classes")
    subplot(122)
    hist(y_pred, bins=len(numpy.unique(y_pred))), title("Distribution over predicted classes"), show()
    
    print_heading("Clusterization metrics:")
    print 'Adjusted Mutual Information', metrics.cluster.adjusted_mutual_info_score(y_true, y_pred)
    print 'Adjusted Rand Score', metrics.cluster.adjusted_rand_score(y_true, y_pred)
    print 'Normalized Mutual Info', metrics.cluster.normalized_mutual_info_score(y_true, y_pred)
    print 'Completeness', metrics.cluster.completeness_score(y_true, y_pred)
    
    matrix = metrics.cluster.contingency_matrix(y_true, y_pred)
    print_heading("Contingency table")

    # print matrix
    pylab.figure(figsize=(18, 6))
    subplot(131), title('Original')
    matrix = numpy.array(matrix, dtype=float)
    pcolor(matrix / matrix.sum(axis=None)), colorbar()
    subplot(132), title('Normed by row')
    pcolor(matrix / matrix.sum(axis=1, keepdims=True)), colorbar()
    subplot(133), title('Normed by column')
    pcolor(matrix / matrix.sum(axis=0, keepdims=True)), colorbar()
    show()
```

# Training AdaBoost classifier
we are using predicted labels of AdaBoost classifier as indices of clusters

```python
%%time
ada = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(min_samples_leaf=50, max_depth=5), n_estimators=70, learning_rate=0.2)
ada_pred = ada.fit(trainX, trainY, sample_weight=trainW).predict(testX)
clusterization_report(testY, ada_pred)
print 'Log likelihood loss:', metrics.log_loss(testY, ada.predict_proba(testX.values))
```

# Training the DBSCAN clusterization
This is the most stable cluterization algorithm in sklearn

```python
%%time
clusterizer = cluster.DBSCAN(leaf_size=30, min_samples=15, eps=1.0)
y_pred = clusterizer.fit_predict(testX.values)
y_pred -= numpy.min(y_pred)
clusterization_report(testY, y_pred)
```
